```python
from deepface import DeepFace
```

```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
import matplotlib.pyplot as plt
```

```python
import urllib.request
```

```python
url = "https://img.mensxp.com/media/content/2020/Aug/header_bannerhappy-girl-taking-a-selfie_5f4a2e57f2699.jpeg?w=1100&h=558&cc=1"
#url = "https://img.etimg.com/thumb/msid-60421376,width-650,height-488,imgsize-44832,,resizemode-75/.jpg"
urllib.request.urlretrieve(url, "image.jpg")
```

```python
image = cv2.imread("image.jpg")
# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the image
plt.imshow(image_rgb)
plt.axis('off')  # Optional: Remove axis ticks and labels
plt.show()
```

```python
prediction = DeepFace.analyze(image)
```
